```python
from sklearn.metrics import accuracy_score,roc_auc_score

def metric(y,pred):
    return roc_auc_score(y,pred,labels=[0,1])
```

```python
from sklearn.model_selection import StratifiedKFold

score_list=[]
test_pred=[]
n_splits=5
kf=StratifiedKFold(n_splits)
score=0
for i,(tr_index,tst_index) in enumerate(kf.split(X_train,y_train)):
    X_train1,y_train1=X_train.loc[tr_index],y_train.loc[tr_index]
    X_test1,y_test1=X_train.loc[tst_index],y_train.loc[tst_index]
    model=CatBoostClassifier(n_estimators=8000,early_stopping_rounds=200,eval_metric='AUC',learning_rate=0.04,use_best_model=True,class_weights=class_weight)
    model.fit(X_train1,y_train1,eval_set=[(X_test1,y_test1)],verbose=100)
    p=model.predict_proba(X_test1)[:,1]
    sc=metric(y_test1,p)
    score_list.append(sc)
    score+=sc/n_splits
    
    pred=model.predict_proba(test_df)[:,1]
    test_pred.append(pred)
    print('Fold {}: {}'.format(i,sc))
print()
print()
print('Avg log: ' ,score)
np.mean(score_list) 
    #print(train_index,test_index,i)
```
